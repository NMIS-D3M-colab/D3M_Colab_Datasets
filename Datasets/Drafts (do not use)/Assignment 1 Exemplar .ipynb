## Step 4 - Data Discovery

### Association Rule Mining

To further dive into this data we will conduct an association rule mining analysis to understand what parts appear in common with which other parts.  In order to do this we will need a dataset in the form:

|        | 100001  | 100006 |100049  |
| ------ | ------- | ------ | ------ | 
| FJELLSE| 100001  |   NaN  | 100049 | 
| HEMNES |  NaN    | 100006 |  Nan   |
|  MALM  | 100001  |   NaN  | 100049 |

To create this matrix we will use the pandas `.mask()` function, and use the row index (part number) as the mask array against all entries that are greater than zero.  The result of this will also be transposed so that the bed assemblies are now the row indicies:

```python
assoc_matrix = joined.mask(joined > 0, joined.index.to_frame(), axis=0).transpose()
assoc_matrix.head()
```

Final pre-processing step we will replace all the zero values in the matrix with `NaN` so they are excluded from the association analysis:

```python
assoc_matrix.replace(to_replace = 0, value = np.nan, inplace=True)
assoc_matrix.head()
```

Install and import libraries to conduct Apriori analysis:

```python
#Install the apyori library 
!pip install apyori --quiet 
from apyori import apriori
```

The pandas dataframe needs to be converted into a raw list of lists to be processed by the apriori function:

```python
transactions = []
for i in range(0, len(assoc_matrix)-1):
  transactions.append([str(assoc_matrix.values[i,j]) for j in range(0, len(assoc_matrix.columns)-1)])
```

Initialise the Apriori alogorithm parameters and put the results into the `results` variable.  The max / min variables have been manually iterated via trial and error to produce a meaningful shortlist of associations.  The main variables that were changed were:

`min_support`: to exclude parts which only appeared on a very small number of the beds (confidence of 0.1 means a part must appear on at least 3 of the 30 bed assemblies in the dataset).

`min_lift`: to highlight associations where the presence of one part indicates a strong likelihood of the other part also being present.

```python
rules = apriori(transactions = transactions, \
                min_support = 0.1, \
                min_confidence = 0.2, \
                min_lift = 5, \
                min_length = 2, \
                max_length = 2)
results = list(rules)
```

The results of the apriori analysis are in a structured python list but this is not readily usable so we define a function `inspect` to process this structure and convert into a pandas dataframe for final investication:

```python
def inspect(results): # define a function for printing out the results
    part1 = [tuple(result[2][0][0])[0] for result in results]
    part2 = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts  = [result[2][0][3] for result in results]
    return list(zip(part1, part2, supports, confidences, lifts))
```

Finally we put the results into a dataframe and apply a sort function to identify those with the highest `lift` values.  With the rule parameters above there are just over 50 associations that meet these critera, so for brevity only the head is shown here: 

```python
AssociatedItems = pd.DataFrame(inspect(results), columns = ['part1', 'part2', 'Support', 'Confidence', 'Lift'])
AssociatedItems.sort_values(by='Lift', ascending=False).head()
```

### Correlation Matrix

As a more visual form of data exploration, we will plot a heatmap showing the similarity between parts on the different beds.  To do this we need a few more plotting libraries:

```python
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

For a quick and simple analysis, we will firstly just plot the correlation matrix that pandas generates automatically from the original joined dataset.

```python
plt.figure(figsize=(18,7))
plt.title("Pandas correlation matrix between bed types")
sns.heatmap(data=joined.corr(), annot=True)
```

This plot makes it very easy to quickly identify which beds have more in common with each other (pale / yellow boxes) as well as beds that have very few parts in common with the others in the catalogue (black horizontal / vertical bands).

This quick analysis could be further extended by just looking at how many parts are shared in common with each other (ignoring quantity).  To do this we apply a mask and transpose the data same as for the association rule mining, but rather than replacing with the part number, we simply set the matrix to a boolean True/False to indicate if that part exists within that bed assembly:

```python
match_matrix = joined.mask(joined > 0, True, axis=0).transpose()
match_matrix.replace(to_replace = 0, value = False, inplace=True)
match_matrix.head()
```

We will now build a match matrix to identify for each combination of beds how many common parts each of them shares.  To do this we will work with a numpy array rather than the dataframe, and will iterate through each row and column and conduct a logical `AND` operation between every pair of rows.  Summing the result of this logical and gives us the total number of parts in common between the two beds.

The results are added to a pair of new arrays - `matches` and `matches_norm`.  The `matches_norm` is normalised by dividing the total number of common parts by the total number of parts in the first assembly.  The diagonal of the matrix is set to NaN as it is irrelevant for this analysis.

```python
parts_array = match_matrix.to_numpy()

matches = []
matches_norm = []

for i in range(0, len(parts_array)):
    current = []
    for j in range(0, len(parts_array)):
        if i != j:
            common_parts = np.logical_and(parts_array[i],parts_array[j]).sum()
            current.append(common_parts)
        else:
            current.append(np.NaN)
    matches.append(current)
    matches_norm.append(np.divide(current, parts_array[i].sum()))
```

The two resulting 2D arrays are then each returned to Pandas dataframes:

```python
matches_df = pd.DataFrame(data = matches, index=match_matrix.index, columns=match_matrix.index)
matches_norm_df = pd.DataFrame(data = matches_norm, index=match_matrix.index, columns=match_matrix.index)
```

Firstly we will plot the two heatmaps:

```python
plt.figure(figsize=(18,7))
plt.title("Common parts between bed assemblies (not normalised)")
sns.heatmap(data=matches_df, annot=True)
```

```python
plt.figure(figsize=(18,7))
plt.title("Common parts between bed assemblies (normalised)")
sns.heatmap(data=matches_norm_df, annot=True)
```

### Part quantity analysis

Applying a similar approach to the previous analysis, as a final exploration we will now identify which parts appear most commonly across the different bed assemblies.  As part of the initial data join we already ran an analysis and identified the two most commonly used parts as `102267` and `105163`.  Here we will look at the pattern of usage of the 5 most common parts across all the bed ranges.

First we will apply similar approach from the end of step 2 above, and then manipulate the dataframe into "long-form" rather than "wide" (pivtot table) form as this is easier to then manipulate the resulting Seaborn plot:

```python
# Use the expression from earlier, and return a series of keys of the top 5 most occuring parts
usage = joined.loc[joined.gt(0).sum(axis=1).sort_values(ascending=False)[0:5].keys()].transpose()
# Append a new column to indicate the type of bed (single or double)
usage['bed_type'] = np.where(usage.index.str.endswith("DB"), "Double", "Guest")
# Name the existing index, so we can refer to it by name in the melt operation
usage.index.names = ['bed_assy']
# Convert the wide format (pivoted) data to long format as preferred by Seaborne library to customise the plot easier
usage = usage.reset_index().melt(id_vars=['bed_assy', 'bed_type'], var_name="part_no", value_name="qty")
usage
```

We will plot these on a swarm plot to visually show the relative proportion of usage of each of these parts:

```python
plt.figure(figsize=(18,12))
plt.xticks(np.arange(0, 21, 2))
plt.xlabel('Quantity required') 
plt.title("Usage quantities of the 5 most commonly occurring parts")
sns.swarmplot(data=usage, x="qty", y="part_no", hue="bed_type")
```
